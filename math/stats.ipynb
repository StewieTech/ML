```python
import scipy.stats as stats

# Define the parameters
n = 10  # Number of trials
p = 1/6  # Probability of success

# Calculate the binomial PMF for each possible outcome (0 to n)
binomial_pmf = [stats.binom.pmf(k, n, p) for k in range(n + 1)]

# Print the probabilities for each outcome
for k, probability in enumerate(binomial_pmf):
    print(f'P(X = {k}) = {probability:.4f}')
```

```python
import matplotlib.pyplot as plt

plt.bar(range(n + 1), binomial_pmf, align='center', alpha=0.5)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xticks(range(n + 1))
plt.show()
```

```python
x = range(n + 1)
binomial_cdf = [stats.binom.cdf(k,n,p) for k in x]

# Create a figure with two subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot the histogram on the first subplot
ax1.bar(x, binomial_pmf, align='center', alpha=0.5)
ax1.set_xlabel('Number of Successes')
ax1.set_ylabel('Probability')
ax1.set_title(f'Binomial PMF (n={n}, p={p})')
ax1.set_xticks(x)

# Plot the CDF on the second subplot
ax2.step(x, binomial_cdf, where='mid')
ax2.set_xlabel('Number of Successes')
ax2.set_ylabel('Cumulative Probability')
ax2.set_title(f'Binomial CDF (n={n}, p={p})')
ax2.set_xticks(x)
ax2.set_ylim(0, 1.1)  # Set the y-axis limits to [0, 1]
ax2.grid(True)

plt.tight_layout()
plt.show()
```
