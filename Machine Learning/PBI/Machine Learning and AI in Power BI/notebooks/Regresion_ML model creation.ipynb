```python
from pycaret.datasets import get_data
dataset = get_data('insurance')
```

```python
data = dataset.sample(frac=0.9, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python

from pycaret.regression import *
exp_reg101 = setup(data = data, target = 'charges', session_id=123,
                  bin_numeric_features = ['age', 'bmi'])
```

```python
best = compare_models(sort = 'R2')
```

```python
tuned_best = tune_model(best)
```

```python

evaluate_model(tuned_best)
```

```python
import matplotlib.pyplot as plt
fig = plt.gcf()
fig.set_size_inches(5, 3.5, forward=True)

plot_model(tuned_best, save=True )
```
